```python
import tkinter
from tkinter import*
from tkinter import ttk
import openpyxl 
from openpyxl import load_workbook

root = Tk()
root.geometry('1000x800')
root.title("College Application Form")

def getvals():
    
    nv=namevalue.get()
    vr=var.get()
    ev=emailvalue.get()
    mv=mobilevalue.get()
    av=addvalue.get()
    sv=signvalue.get()
    mrv=markvalue.get()
    cv=cshvalue.get()
    cbv=comb.get()
    
        
    print("Submitting form")
    print(f"The value of Name is {nv}")
    print(f"The value of Gender is {vr}")
    print(f"The value of Email is {ev}")
    print(f"The value of Mobile is {mv}")
    print(f"The value of Permant Address is {av}")
    print(f"The value of Signature is {sv}")
    print(f"The value of Marks Obtained is {mrv}")
    print(f"The value of Fees Payment is {cv}")
    print(f"The value of Subject is {cbv}")
    
        
    file=('C:\\Users\\Windows SSD\\OneDrive\\Desktop\\Develearn\\PYTHON\\Tkinter GUI\\Collegeform.xlsx')
    workbook=openpyxl.load_workbook(file)
    sheet=workbook.active
    sheet.append([nv,vr,ev,mv,av,sv,mrv,cv,cbv])
    workbook.save(file)
    
submit=Button(root, text='Submit',width=20,bg='blue',fg='white',command=getvals).place(x=240,y=650)

            
f1=Frame(root,bg='black',borderwidth=1,relief=GROOVE)
f1.pack(side=TOP)

L= Label(f1, text="College Application Form",width=20,font='Arial 20 bold')
L.pack(padx=6,pady=10)

L1 = Label(root, text="Name of Student :",width=20,font='Arial 10 bold')
L1.place(x=80,y=130)

namevalue = StringVar()
nameentry = Entry(root,width=25, textvariable=namevalue)
nameentry.place(x=240,y=130)

L2= Label (root, text="Gender :",width=20,font='Arial 10 bold')
L2.place(x=105,y=180)

var = IntVar()

Radiobutton(root, text="Male",padx = 5, variable=var, value=1).place(x=235,y=180)
Radiobutton(root, text="Female",padx = 20, variable=var, value=2).place(x=290,y=180)


L3 = Label(root, text="Email id :",width=20,font='Arial 10 bold')
L3.place(x=100,y=230)

emailvalue = StringVar()
emailentry = Entry(root,width=25, textvariable=emailvalue)
emailentry.place(x=240,y=230)

L4 = Label(root, text="Mobile Number :",width=20,font='Arial 10 bold')
L4.place(x=80,y=280)

mobilevalue = StringVar()
mobileentry = Entry(root,width=25, textvariable=mobilevalue)
mobileentry.place(x=240,y=280)

L5 = Label(root, text="Address :",width=20,font='Arial 10 bold')
L5.place(x=100,y=330)

addvalue = StringVar()
addentry = Entry(root,width=25, textvariable=addvalue)
addentry.place(x=240,y=330)

L6 = Label(root, text="Course Applied for :",font='Arial 10 bold')
L6.place(x=80,y=380)

def item():
    for i in listbox.curselection():
        print(listbox.get(i))

listbox = Listbox(root,height=10,width=25,relief='groove')
listbox.place(x=240,y=380)

listbox.insert(1, 'FYJC ARTS')
listbox.insert(2, 'FYJC COMMERCE')
listbox.insert(3, 'FYJC SCIENCE')
listbox.insert(4, 'SYJC ARTS')
listbox.insert(5, 'SYJC COMMERCE')
listbox.insert(6, 'SYJC SCIENCE')
listbox.insert(7, 'FYBA')
listbox.insert(8, 'SYBA')
listbox.insert(9, 'TYBA')
listbox.insert(10, 'FYCOM')
listbox.insert(11, 'SYCOM')
listbox.insert(12, 'TYCOM')
listbox.insert(13, 'FYBSC')
listbox.insert(14, 'SYBSC')
listbox.insert(15, 'TYBSC')

But=Button (root, text='Select',width=20,bg='green',fg='white',command=item).place(x=240,y=550)

w = Canvas(root, width=300, height=200)
w.place(x=675,y=130)
w.create_rectangle(2, 3, 150, 150, fill="white")

L7 = Label(root, text="Photo",font='Arial 10 bold')
L7.place(x=730,y=290)

L8 = Label(root, text="Signature :",font='Arial 10 bold')
L8.place(x=615,y=330)

signvalue = StringVar()
signentry = Entry(root,width=25, textvariable=signvalue)
signentry.place(x=700,y=330)

L9 = Label(root, text="Percentage :",font='Arial 10 bold')
L9.place(x=600,y=380)

markvalue = StringVar()
markentry = Entry(root,width=25, textvariable=markvalue)
markentry.place(x=700,y=380)

L10 = Label(root, text="Subject :",font='Arial 10 bold')
L10.place(x=625,y=430)

comb=ttk.Combobox(root,width=22,values=['Economics','Maths','Physics','Biology','Engli'])
comb.place(x=700,y=430)

L11 = Label(root, text='Fees payment :',font='Arial 10 bold')
L11.place(x=105,y=600)

cshvalue = IntVar()
csh=Checkbutton(root,text='Cash',font='Arial 10 bold',variable = cshvalue)
csh.place(x=235,y=600)


f2=Frame(root,bg='red',borderwidth=1,relief=GROOVE)
f2.pack(side=BOTTOM)

L12=Label(f2,text= 'For Official use only.',font='Arial 8 bold',bg='light pink')
L12.pack(side=BOTTOM)


root.mainloop()
```
